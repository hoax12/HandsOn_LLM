# Step 1: Phi-3
The first step is to load our model onto the GPU for faster inference. Note that we load the model and tokenizer separately (although that isn't always necessary).

```python
from transformers import AutoModelForCausalLM, AutoTokenizer
#AutoModelForCausalLM: A generic class to load a pre-trained model for causal language modeling (used in text generation tasks).
#AutoTokenizer: A class to load a tokenizer that converts text into a format suitable for the model (e.g., tokens).
# Load model and tokenizer

model = AutoModelForCausalLM.from_pretrained(
    "microsoft/Phi-3-mini-4k-instruct", #This specifies which pre-trained model to load.
    device_map="cuda", #Maps the model to run on GPU if available, for faster computation.
    torch_dtype="auto", #Automatically chooses the appropriate precision (float16 or float32) for efficient computation on the device.
    trust_remote_code = True, #Allows using custom code implementations provided by the model's repository (useful for models requiring custom classes or logic)
    )
tokenizer = AutoTokenizer.from_pretrained("microsoft/Phi-3-mini-4k-instruct")
```

```python
from transformers import pipeline

# Create a pipeline for text generation
generator = pipeline(
    "text-generation",  # Specify the task for the pipeline (text generation in this case)
    model=model,  # Use the pre-trained model loaded earlier
    tokenizer=tokenizer,  # Use the corresponding tokenizer for the model
    return_full_text=False,  # Only return the generated text, not the original prompt
    max_new_tokens=500,  # Limit the generated text to a maximum of 500 tokens
    do_sample=False,  # Disable sampling; use greedy decoding for deterministic output
)
```

```python
# The prompt (user input / query)
messages = [
    {"role": "user", "content": "Provide a detailed receipe to cook chickens."}
    # This dictionary represents the user's prompt.
    # It has a "role" key indicating the sender (user) and a "content" key containing the actual message (the joke request).
]

# Generate output
output = generator(messages)
# This line calls the `generator` pipeline (defined earlier) with the user's prompt to generate a response.
# The output is stored in the `output` variable.
print(output[0]["generated_text"]) #This line prints the generated text from the output.
# It accesses the first element of the `output` list (assuming it's a list) and then retrieves the value associated with the "generated_text" key.
```
